```python
#Question 1: Find Duplicate Elements in a List
lst = [1, 2, 3, 2, 4, 5, 1]

seen = set()
duplicates = set()

for num in lst:
    if num in seen:
        duplicates.add(num)
    else:
        seen.add(num)

print(list(duplicates))
```

```python
#Question 2: List Rotation
lst = [10, 20, 30, 40, 50]
k = 2

k = k % len(lst)   # handle k > length
rotated = lst[-k:] + lst[:-k]

print(rotated)
```

```python
#Question 3: Remove Common Elements from Two Lists
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5]

result = []

for item in list1:
    if item not in list2:
        result.append(item)

print(result)
```

```python
#Question 4: Tuple Element Frequeny
t = (1, 2, 2, 3, 1, 4)

freq = {}

for item in t:
    if item in freq:
        freq[item] += 1
    else:
        freq[item] = 1

print(freq)
```

```python
#Question 5: Convert List to Tuple and Remove Duplicates
lst = [1, 2, 2, 3, 4, 4]

unique = []

for item in lst:
    if item not in unique:
        unique.append(item)

result = tuple(unique)

print(result)
```

```python
#Question 6: Find Maximum and Minimum in Tuple
t = (5, 1, 8, 3)

max_val = t[0]
min_val = t[0]

for num in t:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("Max =", max_val)
print("Min =", min_val)
```

```python
#Question 7: Set Difference and Intersection
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

intersection = set1 & set2
difference = set1 - set2

print("Intersection:", intersection)
print("Difference:", difference)
```

```python
#Question 8: Check Subset
set1 = {1, 2}
set2 = {1, 2, 3, 4}

result = set1.issubset(set2)

print(result)
```
