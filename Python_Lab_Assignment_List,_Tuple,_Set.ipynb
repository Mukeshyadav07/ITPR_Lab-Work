```python
#Question 1: Find Duplicate Elements in a List
lst = [1, 2, 3, 2, 4, 5, 1]

duplicates = []

for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
        if lst[i] == lst[j] and lst[i] not in duplicates:
            duplicates.append(lst[i])

print(duplicates)
```

```python
#Question 2: List Rotation
lst = [10, 20, 30, 40, 50]
k = 2

k = k % len(lst)   # handle k > length
rotated = lst[-k:] + lst[:-k]

print(rotated)
```

```python
#Question 3: Remove Common Elements from Two Lists
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5]

result = []

for item in list1:
    if item not in list2:
        result.append(item)

print(result)
```

```python
#Question 4: Tuple Element Frequeny
t = (1, 2, 2, 3, 1, 4)

freq = {}

for item in t:
    if item in freq:
        freq[item] += 1
    else:
        freq[item] = 1

print(freq)
```

```python
#Question 5: Convert List to Tuple and Remove Duplicates
lst = [1, 2, 2, 3, 4, 4]

unique = []

for item in lst:
    if item not in unique:
        unique.append(item)

result = tuple(unique)

print(result)
```

```python
#Question 6: Find Maximum and Minimum in Tuple
t = (5, 1, 8, 3)

max_val = t[0]
min_val = t[0]

for num in t:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("Max =", max_val)
print("Min =", min_val)
```

```python
#Question 7: Set Difference and Intersection
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

intersection = set1 & set2
difference = set1 - set2

print("Intersection:", intersection)
print("Difference:", difference)
```

```python
#Question 8: Check Subset
set1 = {1, 2}
set2 = {1, 2, 3, 4}

result = set1.issubset(set2)

print(result)
```

```python
#Question 9: Merge and Sort Unique Elements
list1 = [3, 1, 2]
list2 = [2, 4, 5]

merged = list1 + list2
unique = list(set(merged))
unique.sort()

print(unique)
```

```python
#Question 10: Tuple Pair Sum
t = (1, 2, 3, 4, 5)
target = 5

pairs = []

for i in range(len(t)):
    for j in range(i + 1, len(t)):
        if t[i] + t[j] == target:
            pairs.append((t[i], t[j]))

print(pairs)
```

```python
#Question 11: Remove Empty Sets from List
lst = [{1, 2}, set(), {3, 4}, set()]

result = []

for s in lst:
    if s:          # empty set is False
        result.append(s)

print(result)
```

```python
#Question 12: Common Elements in Three Lists
list1 = [1, 2, 3]
list2 = [2, 3, 4]
list3 = [2, 5, 3]

result = []

for item in list1:
    if item in list2 and item in list3:
        if item not in result:
            result.append(item)

print(result)
```

```python
#Question 13: Flatten Nested List
nested = [[1, 2], [3, 4], [5]]

flat = []

for sublist in nested:
    for item in sublist:
        flat.append(item)

print(flat)
```

```python
#Question 14: Symmetric Difference of Sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}

result = set1 ^ set2

print(result)
```

```python
#Question 15: Tuple to Dictionary Mapping
keys = ('a', 'b', 'c')
values = (1, 2, 3)

result = {}

for i in range(len(keys)):
    result[keys[i]] = values[i]

print(result)
```
