```python
#A shopkeeper wants to calculate total bills after discount.
class Shop:
    def __init__(self, price, discount):
        self.price = price
        self.discount = discount

    def final_price(self):
        return self.price - (self.price * self.discount / 100)

s = Shop(3000, 20)
print(s.final_price())
```
